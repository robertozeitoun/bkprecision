```python
import bkprecision as bk
import pylab as pl
import numpy as np
oc=bk.Oscilloscope("192.168.1.167")
```

```python
boc.vdiv(1,'1V')
oc.wfsu(1,4000,0,0)
data=oc.wf(2)
datanormal=(data-np.mean(data))/np.std(data)
pl.figure(figsize=(15,10))
pl.plot(datanormal[0],datanormal[1],markersize=2,label=u"Wave")
pl.legend()
pl.grid()
pl.show()
```
